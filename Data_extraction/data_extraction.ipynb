#### (Bihar, Haryana, Gujarat, Jharkhand, Maharashtra, Telangana,Tripura, Uttar Pradesh and West Bengal)

```python
def get_election_data(url, state_name, constituency_name):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table table-striped table-bordered")

    headers = [header.text.strip() for header in table.find_all('th')]
    desired_headers = ["S.N.", "Candidate", "Party", "EVM Votes", "Postal Votes", "Total Votes", "% of Votes"]
    filtered_headers = [title for title in headers if title in desired_headers]

    rows = []
    for row in table.find_all('tr')[1:]:
        row_data = [cell.text.strip() for cell in row.find_all('td')]
        if len(row_data) == len(filtered_headers):
            rows.append(row_data)

    df = pd.DataFrame(rows, columns=filtered_headers)
    df = df.dropna(how='all')

    df.insert(1, "Constituency", constituency_name)
    df.insert(1, "State", state_name)

    return df

url_n = "https://results.eci.gov.in/AcResultByeJune2024/ConstituencywiseS04195.htm"
state_n = "Bihar"
constituency_n = "Agiaon"
df_main = get_election_data(url_n, state_n, constituency_n)

df_main
```

# General Elections Data:

```python
def get_election_data(url, state_name, party_n):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    table = soup.find("table", class_="table table-striped table-bordered")

    headers = [header.text.strip() for header in table.find_all('th')]
    desired_headers = ['S.No','Parliament Constituency','Winning Candidate','Total Votes','Trailing Candidate','Margin']
    filtered_headers = [title for title in headers if title in desired_headers]

    rows = []
    for row in table.find_all('tr')[1:]:
        row_data = [cell.text.strip() for cell in row.find_all('td')]
        if len(row_data) == len(filtered_headers):
            rows.append(row_data)

    df = pd.DataFrame(rows, columns=filtered_headers)
    df = df.dropna(how='all')

    df.insert(1, "Constituency", party_n)
    df.insert(1, "Party", state_name)

    return df

url_n = "https://results.eci.gov.in/PcResultGenJune2024/partywisewinresult-1745S01.htm"
state_n = "Andhra Pradesh"
party_n = "Telugu Desam"
df_main= get_election_data(url_n, state_n, party_n)

df_main
```
